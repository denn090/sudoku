# Vergleich verschiedener KI-Modelle zur Codegenerierung

## Modelle im Überblick

### ChatGPT-4o

**Stärken:**
- Sehr strukturierte und methodisch fundierte Codegenerierung
- Objektorientierte Ansätze (z. B. Paddle/Ball beim Pong-Spiel)
- Vollständige Umsetzung inkl. Anzeige, Steuerung, Gegner-KI und Punkteanzeige
- Robuste Fehlerbehandlung in Konsolenanwendungen (z. B. Division durch null)
- Merge Sort mit Kommentaren und vielen Testfällen
- Klar verständlicher, gut erklärter Code

**Besonderheiten:**
- Kompatibel mit Jupyter Notebooks (auch bei grafischen Ausgaben)
- Fokus auf Verständlichkeit, ideal für Lernzwecke

---

### ChatGPT o3-mini-high

**Stärken:**
- Präziser, minimalistischer Code – funktional und sauber
- Fokussiert sich strikt auf die Anforderungen im Prompt
- Erweiterungen wie Endscreen beim Pong-Spiel wurden selbstständig eingebaut
- Fehlerbehandlung vorhanden (z. B. division by zero)

**Besonderheiten:**
- Erklärungen nur auf Nachfrage
- Kürzere Ausgaben, aber strukturierter Code
- Gut geeignet für produktive, kompakte Aufgaben

---

### CodeLlama 13B

**Stärken:**
- Solide bei sehr einfachen und kurzen Code-Snippets

**Schwächen:**
- Überfordert bei komplexeren Anforderungen wie Pong
- Keine Interaktivität ohne genaue Vorgaben

**Besonderheiten:**
- Lokal einsetzbar, nützlich für datensensible Anwendungen

---

### DeepSeek Chat

**Stärken:**
- Sehr gute, professionelle Codequalität
- Einhaltung aller Anforderungen auch bei komplexen Projekten
- Optimierte Algorithmen und starkes Exception-Handling
- Exakte Fließkommaberechnungen und gute Randfallbehandlung

**Besonderheiten:**
- Besonders geeignet für produktionsnahe und komplexe Entwicklungsaufgaben

---

### Gemini

**Stärken:**
- Gute Struktur und Stabilität in allen Aufgabenbereichen
- Objektorientierte Umsetzung beim Pong-Spiel
- Sauberer, kommentierter Taschenrechner mit Fehlerbehandlung
- Merge Sort korrekt und nachvollziehbar umgesetzt

**Besonderheiten:**
- Solider Allrounder für Programmieraufgaben in Bildung und Praxis

---

## Fazit (Übersicht)

| Modell            | Codequalität     | Erklärung vorhanden | Komplexität geeignet | Besonderheiten                         |
|------------------|------------------|----------------------|-----------------------|----------------------------------------|
| ChatGPT-4o       | Hoch             | Ja                   | Hoch                  | Ideal für Lernzwecke und Notebooks     |
| o3-mini-high      | Sehr gut         | Nur auf Nachfrage    | Mittel bis Hoch       | Minimalistisch, schnell, kompakt       |
| CodeLlama 13B    | Niedrig          | Nein                 | Nur einfach           | Lokal nutzbar, eingeschränkter Umfang  |
| DeepSeek Chat    | Sehr hoch        | Ja                   | Hoch                  | Für professionelle Anwendungen         |
| Gemini           | Gut              | Ja                   | Mittel                | Solider Allrounder                     |

**Empfehlung nach Einsatzzweck:**
- Für Lehre und saubere Erklärungen: **ChatGPT-4o**, **Gemini**
- Für kurze, präzise Aufgaben: **o3-mini-high**
- Für produktive Entwicklungsarbeit: **DeepSeek Chat**
- Für Offline- oder einfache Szenarien: **CodeLlama 13B**
